```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate synthetic dataset: Temperature prediction based on Time of Day
np.random.seed(42)
X = np.linspace(0, 24, 100)  # Time of day (0 to 24 hours)
y = 20 + 10 * np.sin((X - 6) * np.pi / 12) + np.random.normal(0, 1, X.shape)  # Temperature function with noise


```

```python
def locally_weighted_regression(x0, X, y, tau):
    """ Compute LWR prediction for x0 with bandwidth tau """
    m = X.shape[0]
    X_b = np.column_stack((np.ones(m), X))  # Add bias term
    x0_b = np.array([1, x0])

    # Compute weights using Gaussian kernel
    weights = np.exp(-((X - x0) ** 2) / (2 * tau ** 2))

    # Weighted least squares: theta = (X'WX)^-1 X'W y
    W = np.diag(weights)  # Convert weight vector to diagonal matrix
    theta = np.linalg.inv(X_b.T @ W @ X_b) @ X_b.T @ W @ y  # Normal equation

    return x0_b @ theta  # Prediction for x0
```

```python
 # Generate predictions over the dataset
tau = 2.0  # Bandwidth parameter (controls smoothness)
y_pred = np.array([locally_weighted_regression(x, X, y, tau) for x in X])  # Apply LWR for each point
```

```python
# Plot the results
plt.figure(figsize=(10, 5))
plt.scatter(X, y, label="Actual Data", color="blue", alpha=0.5)
plt.plot(X, y_pred, label=f"LWR Fit (tau={tau})", color="red", linewidth=2)
plt.xlabel("Time of Day (Hours)")
plt.ylabel("Temperature (Â°C)")
plt.title("Locally Weighted Regression - Temperature Prediction")
plt.legend()
plt.show()
```
